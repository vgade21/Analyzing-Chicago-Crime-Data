```python
import pandas as pd
import numpy as np
```

```python
CTA_df = pd.read_csv('C:\\Users\\vamsh\\Documents\\Uni\\UChicago\\Autumn 2022\\MSCA 31012_IP02 - Data Engineering Platforms for Analytics\\Project\\Data\\CTA_System_Information_List_of_L_Stops.csv', encoding = 'gbk', header = 0)
```

```python
CTA_df.head(5)
```

```python
# Adding two empty columns and changing columns headers to upper case for consistency.
CTA_df.columns = CTA_df.columns.str.upper()
CTA_df['LATITUDE'] = 0
CTA_df['LONGITUDE'] = 0
CTA_df.head(5)
```

```python
# Splitting the 'LOCATION' column to pouplate the two new columns
for index,row in enumerate(CTA_df['LOCATION']):
    CTA_df['LATITUDE'][index] = row.replace('(','').replace(')','').split(',')[0]
    CTA_df['LONGITUDE'][index] = row.replace('(','').replace(')','').split(',')[1]

CTA_df.drop(['ADA', 'RED', 'BLUE', 'G', 'BRN', 'P', 'PEXP', 'Y', 'PNK', 'O'], axis = 1, inplace = True)
CTA_df.head(5)
```
